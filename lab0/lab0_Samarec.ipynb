<h1 align="center">Контрольні запитання</h1>

1.Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ SVN та GIT? <br>
   <b> Система контроля версій (СКВ) - це програмне забезпечення, яке призначене для збереження та відстеження змін в файлах і папках проєкту з часом. Головна мета СКВ - забезпечити можливість спільної роботи над проєктами, відстеження історії змін, відмінності між версіями файлів тощо. </b><br>
2.Що відбувається при виконанні команди git add .? <br>
   <b> Команда git add . додає всі зміни в робочій директорії до індексу (staging area), готуючи їх до коміту. Це означає, що Git враховує ці зміни під час наступного коміту.</b><br>
3.Що відбувається при виконанні команди git commit?<br>
   <b> Система контроля версій (СКВ) - це програмне забезпечення, яке призначене для збереження та відстеження змін в файлах і папках проєкту з часом. Головна мета СКВ - забезпечити можливість спільної роботи над проєктами, відстеження історії змін, відмінності між версіями файлів тощо.
</b><br>
4.Якою Git-командою виконується розміщення локальної копії Git-репозиторію проєкта на віддаленому, що знаходиться на GitHub?
<br>
   <b> При виконанні команди git commit Git створює новий коміт (запис) в історії проєкту, який включає в себе всі зміни, які були додані до індексу за допомогою git add. Коміт зазвичай супроводжується коментарем, що пояснює зміни, які були внесені. </b>
